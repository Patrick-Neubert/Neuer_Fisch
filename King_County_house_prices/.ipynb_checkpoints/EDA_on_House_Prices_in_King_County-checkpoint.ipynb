# Exploratory Data Analysis (EDA) on House Prices in King County

## Business Case

**I am looking for investors for my new business idea.**

With my analysis of the house price market of King County, I determined an **indentifier for underestimated houses** in the market.

Further a **location independent predictor for house realistic prices** was found.

The identifier is related to the proportion of the lot compared to the actual house size.
**The bigger the percentage of lot, the smaller the price.**

The predictor relates the price to the house size of the 15 nearest neighbors not to its location directly.
**The bigger the neigbour houses, the higher the price.**

That´s why I came up with the idea to buy underestimated houses with lots bigger or equal to the average house size in King County and to built new houses on the lots.

They should be further selected with the location independent predictor to increase the value as much as possible.

**Short: Buy affordable lots, built new houses with high potential to increase value in the market!**

## Feature Engineering

```python
# Feature: Größe Anwesen --> aus sqft_living + sqft_lot --> estate_size
# Feature: Größe Anwesen der 15 Nachbarn --> sqft_living15 + sqft_lot15
# Feature: Bathrooms/Bedrooms Ratio
# Feature: Zeit seit der letzten Baumaßnahme (Neubau oder Renovierung) = überall wo renovated 0 ist mit yr_built
# auffüllen
# Feature: sqft_living/floors Ratio --> gibt ein Maß für die Durchschnittliche Grundfläche eines Hauses
# hier würde mich auch interessieren, ob die Halbgeschossigen Häuser im Durchschnitt mehr Wohnfläche haben,
# da Halbgeschosse oftmals Speicher sind und nicht zum Wohnraum gehören (sollten)
# Feature: estate_size mit sqft_living/floors Ration --> gibt vielleicht einen besseren Wert
# für die "echte" estate_size wieder
```

### estate_size

```python
house['estate_size'] = (house['sqft_living'] + house['sqft_lot'])
house.head()
```

### estate_size15

```python
house['estate_size15'] = (house['sqft_living15'] + house['sqft_lot15'])
house
```

### Bathrooms/Bedrooms Ratio

```python
house['Bath_Bed_Ratio'] = (house['bathrooms'] / house['bedrooms'])
house
```

### last_construction

```python
# eleminate all NaN in yr_renovated with 0
# yr_renovated_wonan = house.yr_renovated.fillna(value=0)
built = house.yr_built
renovated = house.yr_renovated_wonan

# decision criterium for für combine: take bigger value, if 0
take_smaller = lambda s1, s2: s1 if s1.sum() > s2.sum() else s2
last_construction = renovated.combine(built, take_smaller)
house['last_construction'] = last_construction

house.head()
```

### sqft_living/floors Ratio

```python
house['sqft_living_floors_ratio'] = (house['sqft_living'] / house['floors'])
house
```

### sqft_lot_estate_size ratio

```python
house['sqft_lot_estate_size_ratio'] = (house['sqft_lot'] / house['estate_size'])
house

# percentage of lot of a estate
# is this value big is enough space to built a new house
```

```python
house.info()
```

```python
# date was excluded

x_b = house[['id'
           , 'bedrooms'
           , 'bathrooms'
           , 'sqft_living'
           , 'sqft_lot'
           , 'floors'
           , 'condition'
           , 'grade'
           , 'sqft_above'
           , 'yr_built'
           , 'zipcode'
           , 'lat'
           , 'long'
           , 'sqft_living15'
           , 'sqft_lot15'
           , 'date_encoded'
           , 'waterfront_wonan'
           , 'view_wonan'
           , 'yr_renovated_wonan'
           , 'sqft_basement_float_wonan' 
           , 'estate_size'
           , 'estate_size15'
           , 'Bath_Bed_Ratio'
           , 'last_construction'
           , 'sqft_living_floors_ratio'
           , 'sqft_lot_estate_size_ratio']]
          
x_b = sms.add_constant(x_b)

y_b = house['price']

# Create an OLS model
model_b = sms.OLS(y_b, x_b)
results_b = model_b.fit()
results_b.summary()

# ID has p-Value > 0.05 was taken out
```
